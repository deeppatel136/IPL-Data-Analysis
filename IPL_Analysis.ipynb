```python
df = pd.read_csv('IPL.csv', low_memory=False)
df.head()
df = df.drop(columns=['Unnamed: 0'])
df.columns
df.groupby('batting_team')['team_runs'].sum().sort_values(ascending=False)

import matplotlib.pyplot as plt

team_runs = df.groupby('batting_team')['team_runs'].sum().sort_values(ascending=False)

team_runs.plot(kind='bar')

plt.title("Total Runs by Each Team")
plt.xlabel("Teams")
plt.ylabel("Total Runs")
plt.xticks(rotation=90)
plt.show()

top_batsman = df.groupby('batter')['batter_runs'].sum().sort_values(ascending=False)

top_batsman.head(10)

top_batsman.head(10).plot(kind='bar')

plt.title("Top 10 Batsmen in IPL")
plt.xlabel("Players")
plt.ylabel("Total Runs")
plt.xticks(rotation=45)
plt.show()

df['toss_win_match_win'] = df['toss_winner'] == df['match_won_by']

df['toss_win_match_win'].value_counts()
```
